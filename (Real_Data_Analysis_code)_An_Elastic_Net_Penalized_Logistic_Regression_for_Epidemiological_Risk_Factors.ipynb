```python
# ============================================================
# Real Data Analysis (Colab-ready, copy-paste runnable)
# Proposed: Elastic-Net Penalized Logistic Regression + Bootstrap Stability
# Benchmark: Lasso (L1) Penalized Logistic Regression + same Bootstrap summaries
# Dataset: Breast Cancer Wisconsin (Diagnostic) from scikit-learn
#
# Outputs:
#   outputs_<timestamp>/
#     figs/*.png
#     tables/*.csv (+ *.tex where possible)
#     report_summary.txt
#   outputs_<timestamp>.zip  (auto-download in Colab)
# ============================================================

import os, time, json, shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from dataclasses import dataclass
from typing import Dict, Any, Tuple

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, classification_report

from sklearn.cluster import AgglomerativeClustering

# ---- Colab download helper (safe if not in Colab)
try:
    from google.colab import files
    IN_COLAB = True
except Exception:
    IN_COLAB = False

# -----------------------------
# 0) Reproducibility & outputs
# -----------------------------
SEED = 123
rng_global = np.random.default_rng(SEED)

stamp = time.strftime("%Y%m%d_%H%M%S")
OUTDIR = f"outputs_{stamp}"
FIGDIR = os.path.join(OUTDIR, "figs")
TABDIR = os.path.join(OUTDIR, "tables")
os.makedirs(FIGDIR, exist_ok=True)
os.makedirs(TABDIR, exist_ok=True)

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

def save_fig(filename: str, dpi: int = 200) -> str:
    path = os.path.join(FIGDIR, filename)
    plt.savefig(path, dpi=dpi, bbox_inches="tight")
    return path

def save_table(df: pd.DataFrame, name: str) -> str:
    csv_path = os.path.join(TABDIR, f"{name}.csv")
    tex_path = os.path.join(TABDIR, f"{name}.tex")
    df.to_csv(csv_path, index=False)
    try:
        df.to_latex(tex_path, index=False, float_format="%.6g")
    except Exception:
        pass
    return csv_path

print("Output folder:", OUTDIR)

# -----------------------------
# 1) Load real dataset
# -----------------------------
data = load_breast_cancer()
X = data["data"]
y = data["target"]   # 0=malignant? In sklearn: target_names = ['malignant','benign'] with 0 malignant, 1 benign
feature_names = data["feature_names"]

dfX = pd.DataFrame(X, columns=feature_names)
dfy = pd.Series(y, name="Y")

print("Dataset shape:", dfX.shape)
print("Class balance (Y=1 is benign):", dfy.value_counts(normalize=True).to_dict())

# -----------------------------
# 2) Train/Test split
# -----------------------------
Xtr, Xte, ytr, yte = train_test_split(
    X, y, test_size=0.30, random_state=SEED, stratify=y
)

# -----------------------------
# 3) CV fitters (ElasticNet vs Lasso)
# -----------------------------
def fit_cv(method: str, Xtr: np.ndarray, ytr: np.ndarray, seed: int = 123,
           cv_splits: int = 5) -> Tuple[Pipeline, Dict[str, Any], np.ndarray]:
    """
    method: 'elasticnet' or 'lasso'
    Returns: fitted best pipeline, best_params, coef (p,)
    """
    base = LogisticRegression(
        solver="saga",
        max_iter=8000,
        random_state=seed,
        n_jobs=-1,
        fit_intercept=True
    )

    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", base)
    ])

    Cs = np.logspace(-3, 2, 15)

    if method == "elasticnet":
        pipe.set_params(clf__penalty="elasticnet")
        param_grid = {
            "clf__C": Cs,
            "clf__l1_ratio": [0.2, 0.5, 0.8]
        }
    elif method == "lasso":
        pipe.set_params(clf__penalty="l1")
        param_grid = {
            "clf__C": Cs
        }
    else:
        raise ValueError("method must be 'elasticnet' or 'lasso'")

    cv = StratifiedKFold(n_splits=cv_splits, shuffle=True, random_state=seed)
    gs = GridSearchCV(
        pipe, param_grid=param_grid, scoring="roc_auc",
        cv=cv, n_jobs=-1, refit=True
    )
    gs.fit(Xtr, ytr)

    best = gs.best_estimator_
    coef = best.named_steps["clf"].coef_.ravel().copy()
    return best, gs.best_params_, coef

def selected_set(coef: np.ndarray, thr: float = 1e-8) -> set:
    return set(np.where(np.abs(coef) > thr)[0].tolist())

def jaccard(a: set, b: set) -> float:
    if len(a)==0 and len(b)==0:
        return 1.0
    return len(a & b) / max(1, len(a | b))

# Fit models
pipe_en, best_en, coef_en = fit_cv("elasticnet", Xtr, ytr, seed=SEED)
pipe_l1, best_l1, coef_l1 = fit_cv("lasso", Xtr, ytr, seed=SEED)

# Test AUC + ROC curves
p_en = pipe_en.predict_proba(Xte)[:, 1]
p_l1 = pipe_l1.predict_proba(Xte)[:, 1]
auc_en = roc_auc_score(yte, p_en)
auc_l1 = roc_auc_score(yte, p_l1)

print("\nBest ElasticNet params:", best_en)
print("Test AUC ElasticNet:", auc_en)
print("\nBest Lasso params:", best_l1)
print("Test AUC Lasso:", auc_l1)

# -----------------------------
# 4) Bootstrap stability (fixed hyperparams)
# -----------------------------
def bootstrap_stability_fixed_hyperparams(
    Xtr: np.ndarray, ytr: np.ndarray, fitted_pipe: Pipeline,
    B: int = 300, seed: int = 123
) -> Dict[str, Any]:
    """
    Bootstrap resample train set and refit with the SAME hyperparams as fitted_pipe.
    Returns inclusion freq, sign stability, OR quantiles, Jaccard-to-full distribution.
    """
    rng = np.random.default_rng(seed)
    n, p = Xtr.shape

    full_clf = fitted_pipe.named_steps["clf"]
    full_coef = full_clf.coef_.ravel().copy()
    full_sel = selected_set(full_coef)

    coefs = np.zeros((B, p))
    jac = np.zeros(B)

    # Build params dict from fitted clf
    clf_params = full_clf.get_params()

    for b in range(B):
        idx = rng.integers(0, n, size=n)
        Xb, yb = Xtr[idx], ytr[idx]

        pipe_b = Pipeline([
            ("scaler", StandardScaler()),
            ("clf", LogisticRegression(**clf_params))
        ])
        pipe_b.fit(Xb, yb)

        cb = pipe_b.named_steps["clf"].coef_.ravel()
        coefs[b] = cb
        jac[b] = jaccard(full_sel, selected_set(cb))

    incl = (np.abs(coefs) > 1e-8).mean(axis=0)

    sign_full = np.sign(full_coef)
    # treat zeros in full as 0; sign stability means matching sign exactly
    sign_stab = (np.sign(coefs) == sign_full).mean(axis=0)

    OR = np.exp(coefs)
    OR_med = np.quantile(OR, 0.50, axis=0)
    OR_lo  = np.quantile(OR, 0.025, axis=0)
    OR_hi  = np.quantile(OR, 0.975, axis=0)

    return {
        "full_coef": full_coef,
        "incl_freq": incl,
        "sign_stab": sign_stab,
        "OR_med": OR_med,
        "OR_lo": OR_lo,
        "OR_hi": OR_hi,
        "jac_to_full": jac
    }

B_BOOT = 300
stab_en = bootstrap_stability_fixed_hyperparams(Xtr, ytr, pipe_en, B=B_BOOT, seed=SEED+101)
stab_l1 = bootstrap_stability_fixed_hyperparams(Xtr, ytr, pipe_l1, B=B_BOOT, seed=SEED+202)

# -----------------------------
# 5) Optional: correlation clusters (to illustrate grouping)
# -----------------------------
# cluster features based on absolute correlation (distance = 1 - |corr|)
corr = dfX.corr().values
dist = 1.0 - np.abs(corr)
# Agglomerative clustering on precomputed distances
n_clusters = 8  # tweak if you want more/less granularity
clust = AgglomerativeClustering(
    n_clusters=n_clusters,
    metric="precomputed",
    linkage="average"
)
cluster_id = clust.fit_predict(dist)

# -----------------------------
# 6) Build tables (copyable + saved)
# -----------------------------
def make_feature_table(method_name: str, stab: Dict[str, Any], best_params: Dict[str, Any]) -> pd.DataFrame:
    full_coef = stab["full_coef"]
    df = pd.DataFrame({
        "feature": feature_names,
        "cluster": cluster_id,
        "coef_full": full_coef,
        "OR_median": stab["OR_med"],
        "OR_2.5%": stab["OR_lo"],
        "OR_97.5%": stab["OR_hi"],
        "incl_freq": stab["incl_freq"],
        "sign_stab": stab["sign_stab"],
        "abs_coef_full": np.abs(full_coef),
    })
    df["selected_full"] = (df["abs_coef_full"] > 1e-8).astype(int)
    df["method"] = method_name
    df["best_params"] = json.dumps(best_params)
    return df.sort_values(["incl_freq", "sign_stab", "abs_coef_full"], ascending=False).reset_index(drop=True)

feat_en = make_feature_table("ElasticNet", stab_en, best_en)
feat_l1 = make_feature_table("Lasso", stab_l1, best_l1)

top20_en = feat_en.head(20).copy()
top20_l1 = feat_l1.head(20).copy()

# Performance summary table
sel_en = selected_set(coef_en)
sel_l1 = selected_set(coef_l1)
overlap = len(sel_en & sel_l1)

perf = pd.DataFrame([
    {
        "method": "ElasticNet",
        "test_auc": auc_en,
        "n_selected_full": len(sel_en),
        "mean_bootstrap_jaccard_to_full": float(np.mean(stab_en["jac_to_full"])),
        "sd_bootstrap_jaccard_to_full": float(np.std(stab_en["jac_to_full"])),
        "best_params": json.dumps(best_en)
    },
    {
        "method": "Lasso",
        "test_auc": auc_l1,
        "n_selected_full": len(sel_l1),
        "mean_bootstrap_jaccard_to_full": float(np.mean(stab_l1["jac_to_full"])),
        "sd_bootstrap_jaccard_to_full": float(np.std(stab_l1["jac_to_full"])),
        "best_params": json.dumps(best_l1)
    },
    {
        "method": "Overlap(EN∩Lasso)",
        "test_auc": np.nan,
        "n_selected_full": overlap,
        "mean_bootstrap_jaccard_to_full": np.nan,
        "sd_bootstrap_jaccard_to_full": np.nan,
        "best_params": ""
    }
])

# Show + save tables
print("\n=== PERFORMANCE SUMMARY (copy-ready) ===")
display(perf)
print(perf.to_markdown(index=False))
save_table(perf, "performance_summary")

print("\n=== TOP 20 STABLE FEATURES: ElasticNet (copy-ready) ===")
display(top20_en)
print(top20_en.to_markdown(index=False))
save_table(top20_en, "top20_stability_elasticnet")

print("\n=== TOP 20 STABLE FEATURES: Lasso (copy-ready) ===")
display(top20_l1)
print(top20_l1.to_markdown(index=False))
save_table(top20_l1, "top20_stability_lasso")

# Save full feature-level tables
save_table(feat_en, "feature_stability_full_elasticnet")
save_table(feat_l1, "feature_stability_full_lasso")

# -----------------------------
# 7) Plots (all shown + saved)
# -----------------------------
# Plot A: Predictor correlation heatmap (to visualize multicollinearity)
plt.figure(figsize=(7,6))
plt.imshow(corr, aspect="auto")
plt.colorbar(label="Correlation")
plt.title("Breast Cancer predictors: correlation heatmap")
plt.xlabel("Features")
plt.ylabel("Features")
save_fig("corr_heatmap.png")
plt.show()

# Plot B: ROC curves
fpr_en, tpr_en, _ = roc_curve(yte, p_en)
fpr_l1, tpr_l1, _ = roc_curve(yte, p_l1)

plt.figure()
plt.plot(fpr_en, tpr_en, label=f"ElasticNet (AUC={auc_en:.3f})")
plt.plot(fpr_l1, tpr_l1, label=f"Lasso (AUC={auc_l1:.3f})")
plt.plot([0,1],[0,1], linestyle="--", linewidth=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC curves on test set")
plt.legend()
save_fig("roc_curves.png")
plt.show()

# Plot C: Bootstrap Jaccard distributions
plt.figure()
plt.hist(stab_en["jac_to_full"], bins=20, alpha=0.7, label="ElasticNet")
plt.hist(stab_l1["jac_to_full"], bins=20, alpha=0.7, label="Lasso")
plt.xlabel("Jaccard(bootstrap selection, full selection)")
plt.ylabel("Count")
plt.title(f"Bootstrap selection stability (B={B_BOOT})")
plt.legend()
save_fig("bootstrap_jaccard_hist.png")
plt.show()

# Plot D: Top-20 inclusion frequency comparison (EN vs Lasso)
k = 20
top_feats = list(top20_en["feature"].values)
idx_map = {f:i for i,f in enumerate(feature_names)}
top_idx = [idx_map[f] for f in top_feats]

plt.figure(figsize=(10,4))
x = np.arange(k)
plt.bar(x - 0.2, stab_en["incl_freq"][top_idx], width=0.4, label="ElasticNet")
plt.bar(x + 0.2, stab_l1["incl_freq"][top_idx], width=0.4, label="Lasso")
plt.xticks(x, top_feats, rotation=75, ha="right")
plt.ylabel("Bootstrap inclusion frequency")
plt.title("Top stable features (from ElasticNet): inclusion frequency comparison")
plt.legend()
save_fig("top20_inclusion_freq_compare.png")
plt.show()

# Plot E: Odds-ratio intervals for Top-15 (ElasticNet)
k_or = 15
top15 = top20_en.head(k_or).copy()
ypos = np.arange(k_or)

OR_med = top15["OR_median"].values
OR_lo  = top15["OR_2.5%"].values
OR_hi  = top15["OR_97.5%"].values

plt.figure(figsize=(8,4.8))
plt.errorbar(OR_med, ypos, xerr=[OR_med-OR_lo, OR_hi-OR_med], fmt="o")
plt.axvline(1.0, linestyle="--", linewidth=1)
plt.yticks(ypos, top15["feature"].values)
plt.xlabel("Odds Ratio (median and 95% bootstrap interval)")
plt.title("ElasticNet: Top stable features (OR intervals)")
save_fig("elasticnet_top15_or_intervals.png")
plt.show()

# Plot F: Coefficient magnitude barplot (ElasticNet, top 20 by |coef|)
coef_df = pd.DataFrame({
    "feature": feature_names,
    "coef_full": stab_en["full_coef"],
    "abs_coef": np.abs(stab_en["full_coef"])
}).sort_values("abs_coef", ascending=False).head(20)

plt.figure(figsize=(10,4))
plt.bar(np.arange(len(coef_df)), coef_df["coef_full"].values)
plt.xticks(np.arange(len(coef_df)), coef_df["feature"].values, rotation=75, ha="right")
plt.ylabel("Coefficient (full fit, standardized X)")
plt.title("ElasticNet: Top 20 coefficients by absolute magnitude")
save_fig("elasticnet_top20_coef_bar.png")
plt.show()

# -----------------------------
# 8) Extra: simple classification report at default threshold
# -----------------------------
# sklearn target: 1=benign, 0=malignant. Use threshold 0.5 on P(Y=1).
yhat_en = (p_en >= 0.5).astype(int)
yhat_l1 = (p_l1 >= 0.5).astype(int)

cm_en = confusion_matrix(yte, yhat_en)
cm_l1 = confusion_matrix(yte, yhat_l1)

cm_df = pd.DataFrame({
    "method": ["ElasticNet", "Lasso"],
    "TN": [cm_en[0,0], cm_l1[0,0]],
    "FP": [cm_en[0,1], cm_l1[0,1]],
    "FN": [cm_en[1,0], cm_l1[1,0]],
    "TP": [cm_en[1,1], cm_l1[1,1]],
})
print("\n=== CONFUSION MATRICES @ 0.5 threshold (copy-ready) ===")
display(cm_df)
print(cm_df.to_markdown(index=False))
save_table(cm_df, "confusion_matrices_threshold_0p5")

# -----------------------------
# 9) Write a short text summary (saved)
# -----------------------------
summary_txt = []
summary_txt.append("Real-data analysis: Breast Cancer Wisconsin (Diagnostic)\n")
summary_txt.append(f"Timestamp: {stamp}\n")
summary_txt.append("Methods: ElasticNet logistic vs Lasso logistic (both CV-tuned), + bootstrap stability.\n")
summary_txt.append(f"ElasticNet best params: {best_en}\n")
summary_txt.append(f"Lasso best params: {best_l1}\n")
summary_txt.append(f"Test AUC ElasticNet: {auc_en:.6f}\n")
summary_txt.append(f"Test AUC Lasso:      {auc_l1:.6f}\n")
summary_txt.append(f"Selected features (ElasticNet): {len(sel_en)}\n")
summary_txt.append(f"Selected features (Lasso):      {len(sel_l1)}\n")
summary_txt.append(f"Overlap selected:               {overlap}\n")
summary_txt.append(f"Mean bootstrap Jaccard (ElasticNet): {np.mean(stab_en['jac_to_full']):.6f}\n")
summary_txt.append(f"Mean bootstrap Jaccard (Lasso):      {np.mean(stab_l1['jac_to_full']):.6f}\n")

with open(os.path.join(OUTDIR, "report_summary.txt"), "w") as f:
    f.write("".join(summary_txt))

# -----------------------------
# 10) Zip outputs + download (Colab)
# -----------------------------
zip_path = shutil.make_archive(OUTDIR, "zip", OUTDIR)
print("\nZipped outputs to:", zip_path)

if IN_COLAB:
    files.download(zip_path)
```
